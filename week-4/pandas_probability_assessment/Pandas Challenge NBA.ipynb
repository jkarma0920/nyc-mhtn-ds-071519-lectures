```python
df = pd.DataFrame()

for person in data:
    if len(person['GAMELOG']) == 0:
        continue
    df1 = pd.DataFrame(person, columns=['DISPLAY_FIRST_LAST', 'PERSON_ID', 'TEAM_ID'], index=[0])
    df2 = pd.DataFrame(person['GAMELOG'])
    merged = pd.merge(df1, df2, left_on='PERSON_ID', right_on='Player_ID', how='outer')
    df = pd.concat([df, merged])
    
df.reset_index(drop=True, inplace=True)
```

## 9:

What is the probability that a randomly selected player from last season would average more than 20 points per game?

```python
# Sean's code
player_grouped = df.groupby['DISPLAY_FIRST_LAST']
np.sum(player_grouped['PTS'].mean()>20)/530
```

```python
# Rafi's code
import statistics

def group_players(df):
    return df.groupby(['DISPLAY_FIRST_LAST'])

#retrieve total number of unique players
def total_players(df):
    grouped_players = group_players(df)
    return len(grouped_players)
#calculate odds of getting num1 in sample space num2
def calculate_odds(num1, num2):
    return (num1/num2)

#return a DF of all the games featuring players who have scored over 20
def attribute_over_20(df, attribute):
    grouped_players = group_players(df)
    #filter by selecting where mean of attribute is greater than 20
    over_20_avg = grouped_players.filter(lambda x : statistics.mean(x[attribute]) > 20)
    return over_20_avg
def num_of_20_ppg(df):
    #get data_frame of all games with players who have 20 or better average
    players_over_20 = attribute_over_20(df, 'PTS')
    grouped_players = group_players(players_over_20)
    #regroup and get the length to get count
    num_of_players = len(grouped_players)
    return num_of_players
#probability of getting a player with 20 plus ppg average at random
def probs_of_20_avg(df):
    players_total = total_players(df)
    over_20_ppg_players = num_of_20_ppg(df)
    return calculate_odds(over_20_ppg_players, players_total)
```
